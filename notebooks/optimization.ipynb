```python
def objective(x):
    return (1 - x[0])**2 + 100 * (x[1] - x[0]**2)**2
```

```python
from cobra.core.optimizers.gradient.gd import GradientDescentOptimizer


opt = GradientDescentOptimizer(
    learning_rate=0.001,
    max_iter=1000,
    verbose=True
)

x0 = [0.0, 0.0]
result = opt(objective, x0)
```

```python
import numpy as np

from cobra.core.optimizers.search.grid import GridSearchOptimizer


grid = GridSearchOptimizer(param_grid={'x': np.linspace(-5, 5, 100), 'y': np.linspace(-5, 5, 100)}, verbose=True)
result = grid(objective)
```
